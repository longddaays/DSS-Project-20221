# Topsis Model

```python
import pandas as pd
from topsis import Normalize, Calc_Values

def topsis_pipy(temp_dataset, dataset, nCol, weights, impact):
    
    temp_dataset = Normalize(temp_dataset, nCol, weights)

    
    p_sln, n_sln = Calc_Values(temp_dataset, nCol, impact)

    
    
    score = []
    for i in range(len(temp_dataset)):
        temp_p, temp_n = 0, 0
        for j in range(1, nCol):
            temp_p = temp_p + (p_sln[j-1] - temp_dataset.iloc[i, j])**2
            temp_n = temp_n + (n_sln[j-1] - temp_dataset.iloc[i, j])**2
        temp_p, temp_n = temp_p**0.5, temp_n**0.5
        score.append(temp_n/(temp_p + temp_n))
    dataset['topsis_score'] = score

    
    dataset['rank'] = (dataset['topsis_score'].rank(
        method='max', ascending=False))
    dataset = dataset.astype({"rank": int})
```

# Topsis Ranking

```python
data = pd.read_csv('../data/data_time_transformed.csv')
# dataa = newData(data.shape[0])
data.head(2)
```

```python
def topsis_ranking(weights,n):    
    dataa = newData(n)
    dat = dataa.copy()
    temp = dataa.copy()
    # weights = [10,5,1,2]
    impact = ["+","-","+","-"]
    topsis_pipy(dat,temp,4,weights,impact)
    temp[["name","dishes"]] = temp["name"].str.split(';', 1, expand=True)
    return temp.sort_values(by=['rank'])
```
