# Topsis Model

```python
import pandas as pd
from topsis import Normalize, Calc_Values

def topsis_pipy(temp_dataset, dataset, nCol, weights, impact):
    
    temp_dataset = Normalize(temp_dataset, nCol, weights)

    
    p_sln, n_sln = Calc_Values(temp_dataset, nCol, impact)

    
    
    score = []
    for i in range(len(temp_dataset)):
        temp_p, temp_n = 0, 0
        for j in range(1, nCol):
            temp_p = temp_p + (p_sln[j-1] - temp_dataset.iloc[i, j])**2
            temp_n = temp_n + (n_sln[j-1] - temp_dataset.iloc[i, j])**2
        temp_p, temp_n = temp_p**0.5, temp_n**0.5
        score.append(temp_n/(temp_p + temp_n))
    dataset['topsis_score'] = score

    
    dataset['rank'] = (dataset['topsis_score'].rank(
        method='max', ascending=False))
    dataset = dataset.astype({"rank": int})
```

# Restaurant Recommendations based on Users' past experience

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def dish_recommendations(dish, n):

    #Vec-tơ hóa từ
    vectorizer = TfidfVectorizer(analyzer = 'word', ngram_range = (1,2), min_df = 0)
    vectorizer.fit(temp['cuisines'])
    x_train_tfidf = vectorizer.transform(temp['cuisines'])

    #Độ tương đồng cosine
    cosine_sim = cosine_similarity(x_train_tfidf, x_train_tfidf)

    weights = input([])
    temp = topsis_ranking(weights)
    dishes = temp['cuisines']
    indices = pd.Series(temp.index, index = temp['cuisines']) 
    idx = indices[dish]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key = lambda x:x[1], reverse = True)
    sim_scores = sim_scores[1:n+1]
    dish_indices = [i[0] for i in sim_scores]
    
    return temp['name'].iloc[dish_indices]

def restaurant_recom(n):
    result = pd.DataFrame()
    for i in temp['cuisines']:
        output = pd.DataFrame(dish_recommendations(i,n))
        result = result.append(output)
    
    return result['name'].value_counts()
```
